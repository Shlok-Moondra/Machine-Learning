## Calculating the expenses

```python
print("Profit Calculation:", profit_calculation)
```

```python
def  calculate_profit_revenue(revenue,cost):
    profit = revenue-cost
    margin = (profit*100)/revenue
    return profit,margin
```

```python
revenue = 100
cost = 37

profit,margin = calculate_profit_revenue(revenue,cost)
print("Profit:", profit)
print("Margin:", margin)
```

```python
response = calculate_profit_revenue(200, 50)
type(response)
```

```python
def calculate_multiple_profits(*args):
    total_profit = sum(args)
    return total_profit
calculate_multiple_profits(100, 200, 300, 400)
```
