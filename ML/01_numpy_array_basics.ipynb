# NumPy Array Basics

Basic NumPy arrays, array creation, attributes, reshaping, and a simple element check.

```python
# 1. Import statements
import numpy as np
```

```python
# 2. Getting data / creating arrays
vector = np.array([1, 2, 3])
matrix = np.array([[1, 2, 3], [4, 5, 6]])
zeros = np.zeros((3, 3))
ones = np.ones((2, 2))
constant = np.full((2, 2), 7)
random_values = np.random.default_rng(42).random((3, 3))

print('Vector:', vector)
print('Matrix:\n', matrix)
print('Zeros:\n', zeros)
print('Ones:\n', ones)
print('Constant values:\n', constant)
print('Random values:\n', random_values)
```

```python
# 3. Inspecting array structure
sample = np.array([[1, 2, 3], [4, 5, 6]])

print('Shape:', sample.shape)
print('Dimensions:', sample.ndim)
print('Total values:', sample.size)
print('Data type:', sample.dtype)
print('Bytes per item:', sample.itemsize)
print('Transposed array:\n', sample.T)
```

```python
# 4. Reshaping arrays
numbers = np.arange(12)
reshaped = numbers.reshape(3, 4)

print('Original array:', numbers)
print('Reshaped array:\n', reshaped)
```

```python
# 5. Testing against one datapoint / element
row_index = 1
column_index = 2
datapoint = reshaped[row_index, column_index]

print(f'Value at row {row_index}, column {column_index}:', datapoint)
```
